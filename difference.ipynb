```python
# Write a simple code that shows the difference between matplotlib and seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Generate a sample dataset
np.random.seed(42)
data = {
    "x": np.random.rand(50),
    "y": np.random.rand(50),
    "category": np.random.choice(['A', 'B'], size=50)
}
df = pd.DataFrame(data)

# Matplotlib: Manual Plotting
plt.figure(figsize=(10, 5))

# Subplot 1: Matplotlib
plt.subplot(1, 2, 1)
for category in df['category'].unique():
    subset = df[df['category'] == category]
    plt.scatter(subset['x'], subset['y'], label=f'Category {category}')
plt.title("Scatter Plot (Matplotlib)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Subplot 2: Seaborn
plt.subplot(1, 2, 2)
sns.scatterplot(
    data=df, x='x', y='y', hue='category', palette='coolwarm', style='category'
)
plt.title("Scatter Plot (Seaborn)")

plt.tight_layout()
plt.show()
```

```python
# Generate some sample data
np.random.seed(0)
data = np.random.randn(100)

# Matplotlib Example: Creating a histogram
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.hist(data, bins=20, color='blue', alpha=0.7)
plt.title('Histogram Using Matplotlib')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Seaborn Example: Creating a histogram
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
sns.histplot(data, bins=20, color='blue', kde=True)
plt.title('Histogram Using Seaborn')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Generate sample data
np.random.seed(0)
data = pd.DataFrame({
    'x': np.arange(1, 11),
    'y': np.random.randint(1, 20, size=10)
})

# Matplotlib Example: Creating a bar plot with multiple lines of code
plt.figure(figsize=(10, 5))
plt.bar(data['x'], data['y'], color='blue')
plt.title('Bar Plot Using Matplotlib')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(data['x'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Seaborn Example: Creating the same bar plot with fewer lines of code
plt.figure(figsize=(10, 5))
sns.barplot(x='x', y='y', data=data, color='blue')
plt.title('Bar Plot Using Seaborn')
plt.show()
```

### Advantage of using matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
np.random.seed(0)
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create a figure with multiple subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

# Plot 1: Sine Wave
axs[0].plot(x, y1, color='blue', linewidth=2)
axs[0].set_title('Sine Wave', fontsize=16)
axs[0].set_xlabel('X-axis', fontsize=14)
axs[0].set_ylabel('sin(x)', fontsize=14)
axs[0].grid(True)
axs[0].annotate('Max Value', xy=(1.57, 1), xytext=(2, 1.5),
                 arrowprops=dict(facecolor='black', shrink=0.05))

# Plot 2: Cosine Wave
axs[1].plot(x, y2, color='orange', linewidth=2)
axs[1].set_title('Cosine Wave', fontsize=16)
axs[1].set_xlabel('X-axis', fontsize=14)
axs[1].set_ylabel('cos(x)', fontsize=14)
axs[1].grid(True)
axs[1].annotate('Max Value', xy=(0, 1), xytext=(2, 1.5),
                 arrowprops=dict(facecolor='black', shrink=0.05))

# Plot 3: Tangent Wave
axs[2].plot(x, y3, color='green', linewidth=2)
axs[2].set_title('Tangent Wave', fontsize=16)
axs[2].set_xlabel('X-axis', fontsize=14)
axs[2].set_ylabel('tan(x)', fontsize=14)
axs[2].set_ylim(-10, 10)  # Limit y-axis for better visibility
axs[2].grid(True)

# Add a main title for the entire figure
plt.suptitle('Trigonometric Functions', fontsize=20)

# Adjust layout to prevent overlap
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Generate sample data
np.random.seed(0)
data = pd.DataFrame({
    'x': np.linspace(1, 10, 100),
    'y1': np.sin(np.linspace(1, 10, 100)),
    'y2': np.cos(np.linspace(1, 10, 100)),
    'y3': np.tan(np.linspace(1, 10, 100))
})

# Matplotlib Example: Creating a detailed line plot with customization
plt.figure(figsize=(12, 8))

# Plot sine wave
plt.plot(data['x'], data['y1'], label='Sine', color='blue', linewidth=2)
plt.plot(data['x'], data['y2'], label='Cosine', color='orange', linewidth=2)
plt.plot(data['x'], data['y3'], label='Tangent', color='green', linewidth=2)

# Customizing the plot
plt.title('Trigonometric Functions', fontsize=20)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.ylim(-10, 10)  # Limit y-axis for better visibility
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(True)
plt.legend()
plt.show()

# Seaborn Example: Creating a similar plot with less customization
plt.figure(figsize=(12, 8))
sns.lineplot(data=data[['x', 'y1', 'y2']], x='x', y='y1', label='Sine', color='blue')
sns.lineplot(data=data[['x', 'y2']], x='x', y='y2', label='Cosine', color='orange')
sns.lineplot(data=data[['x', 'y3']], x='x', y='y3', label='Tangent', color='green')

# Adding title and labels in Seaborn
plt.title('Trigonometric Functions (Seaborn)', fontsize=20)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.ylim(-10, 10)  # Limit y-axis for better visibility
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(True)
plt.legend()
plt.show()
```
